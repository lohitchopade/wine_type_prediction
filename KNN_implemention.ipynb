```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
df=pd.read_csv("wine.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.groupby('type').groups.keys()
```

```python
df["type"].replace({'red':0, 'white':1},inplace=True)
```

```python
###df.fillna({'fixed acidity':df['fixed acidity'].mean, 'volatile acidity':df['volatile acidity'].mean, 'citric acid': df['citric acid'].mean, 'residual sugar':df['residual sugar'].mean, 'chlorides':df['chlorides'].mean, 'pH':df['pH'].mean, 'sulphates':df['sulphates'].mean},inplace=True)

df.dropna(inplace=True)
```

```python
df.info()
```

# normalization

```python
x=df.drop('quality',axis=1)
normal_scale=MinMaxScaler()
normal_scale.fit(x)
normal_array=normal_scale.transform(x)

x_df=pd.DataFrame(normal_array,columns=x.columns)
```

```python
x_df.describe()
```

# Standardization

```python
standard_scale=StandardScaler()
standard_scale.fit(x_df)
std_array=standard_scale.transform(x_df)
x_df1=pd.DataFrame(std_array,columns=x.columns)
x_df1.describe()
```

```python
x=x_df1
y=df['quality']

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=10,stratify=y)
x_train
```

# KNN model

```python
knn_model=KNeighborsClassifier(n_neighbors=5,p=2)
knn_model.fit(x_train,y_train)
```

```python
y_pred=knn_model.predict(x_test)
y_pred
```

```python
conf_mat=confusion_matrix(y_test,y_pred)
print(conf_mat)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Hyperparameter Tunninng

```python
knn_model=KNeighborsClassifier(n_neighbors=5,p=2)
hyperparameters={'n_neighbors':np.arange(2,30),'p':[1,2]}

GSCV_knn=GridSearchCV(knn_model, hyperparameters, cv=5)
GSCV_knn.fit(x_train,y_train)
```

```python
GSCV_knn.best_estimator_
```

```python
GSCV_knn.best_params_
```

```python
knn=GSCV_knn.best_estimator_
knn.fit(x_train,y_train)
```

```python
y_pred_after_hypT=knn.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_after_hypT))
```

```python
accuracy_score(y_test,y_pred_after_hypT)
```
